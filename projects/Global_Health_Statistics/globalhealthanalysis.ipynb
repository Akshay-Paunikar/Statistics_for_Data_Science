### Global Health Statistics

Global Health Statistics: Disease Prevalence, Treatments & Outcomes by Country

#### About Dataset:

This dataset provides comprehensive statistics on global health, focusing on various diseases, treatments, and outcomes. The data spans multiple countries and years, offering valuable insights for health research, epidemiology studies, and machine learning applications. The dataset includes information on the prevalence, incidence, and mortality rates of major diseases, as well as the effectiveness of treatments and healthcare infrastructure.

#### Dataset Use Cases:

This dataset can be used for:

 - Healthcare Policy Analysis: Understanding which diseases are most prevalent and which countries require more investment in healthcare infrastructure.
 - Epidemiological Studies: Studying the correlation between disease prevalence and socio-economic factors like income, education, and urbanization.
 - Machine Learning Models: Training predictive models to forecast disease trends, mortality rates, and treatment effectiveness based on historical data.
 - Global Health Research: Identifying regions that need targeted interventions or public health campaigns.

##### Column Descriptions:

 - <b>Country:</b> The name of the country where the health data was recorded.

 - <b>Year:</b> The year in which the data was collected.

 - <b>Disease Name:</b> The name of the disease or health condition tracked.

 - <b>Disease Category:</b> The category of the disease (e.g., Infectious, Non-Communicable).

 - <b>Prevalence Rate (%):</b> The percentage of the population affected by the disease.

 - <b>Incidence Rate (%):</b> The percentage of new or newly diagnosed cases.

 - <b>Mortality Rate (%):</b> The percentage of the affected population that dies from the disease.

 - <b>Age Group:</b> The age range most affected by the disease.

 - <b>Gender:</b> The gender(s) affected by the disease (Male, Female, Both).

 - <b>Population Affected:</b> The total number of individuals affected by the disease.

 - <b>Healthcare Access (%):</b> The percentage of the population with access to healthcare.

 - <b>Doctors per 1000:</b> The number of doctors per 1000 people.

 - <b>Hospital Beds per 1000:</b> The number of hospital beds available per 1000 people.

 - <b>Treatment Type:</b> The primary treatment method for the disease (e.g., Medication, Surgery).

 - <b>Average Treatment Cost (USD):</b> The average cost of treating the disease in USD.

 - <b>Availability of Vaccines/Treatment:</b> Whether vaccines or treatments are available.

 - <b>Recovery Rate (%):</b> The percentage of people who recover from the disease.

 - <b>DALYs:</b> Disability-Adjusted Life Years, a measure of disease burden.

 - <b>Improvement in 5 Years (%):</b> The improvement in disease outcomes over the last five years.

 - <b>Per Capita Income (USD):</b> The average income per person in the country.

 - <b>Education Index:</b> The average level of education in the country.

 - <b>Urbanization Rate (%):</b> The percentage of the population living in urban areas.

```python
# Let's visualize the categorical data
plt.figure(figsize=(15, 6))

# Disease Names & their counts

ax1 = plt.subplot2grid((1,2),(0,0))
sns.countplot(y='Disease Name', data=ghs, ax=ax1, order=ghs['Disease Name'].value_counts().index, palette='copper')
plt.title('Disease Names & their counts')
plt.xlabel('Counts')
plt.ylabel('Disease Names')

# Disease Categories & their counts

ax2 = plt.subplot2grid((1,2),(0,1))
sns.countplot(y='Disease Category', data=ghs, ax=ax2, order=ghs['Disease Category'].value_counts().index, palette='copper')
plt.title('Disease Categories & their counts')
plt.xlabel('Counts')
plt.ylabel('Disease Categories')

plt.tight_layout()
plt.show()
```

```python
# Age Groups in the dataset
ghs['Age Group'].value_counts()
```

```python
# Gender in the dataset
ghs['Gender'].value_counts()
```

```python
plt.figure(figsize=(15, 4))

# Gender distribution pie chart

ax1 = plt.subplot2grid((1,2),(0,0))
gender_counts = ghs['Gender'].value_counts()

plt.pie(gender_counts, labels=gender_counts.index, autopct=lambda p: f'{p:.2f}%\n({int(p * sum(gender_counts) / 100)})', 
        startangle=90, colors=sns.color_palette('crest'))

plt.title('Gender Distribution')

# Age Group distribution pie chart

ax2 = plt.subplot2grid((1,2),(0,1))
age_counts = ghs['Age Group'].value_counts()
plt.pie(age_counts, labels=age_counts.index, autopct=lambda p: f'{p:.2f}%\n({int(p * sum(age_counts) / 100)})', 
        startangle=90, colors=sns.color_palette('crest'))
plt.title('Age Group Distribution')

plt.tight_layout()
plt.show()
```
