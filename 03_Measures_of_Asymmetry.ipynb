### What are the measures of asymmetry in statistics?

Measures of asymmetry in statistics are used to describe the degree to which a distribution or dataset is skewed, or not symmetrical. The two most commonly used measures of asymmetry are skewness and kurtosis.

 - Skewness: Skewness is a measure of the asymmetry of a distribution, and it indicates the degree to which the data is skewed to the left or right.
```python
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# create a sample dataset
data = np.random.normal(0, 1, 1000)

# calculate the skewness and kurtosis
skewness = skew(data)
kurt = kurtosis(data)

# plot the distribution
sns.histplot(data, kde=True)
plt.title("Distribution of Data")

# add labels to the plot
plt.text(3, 70, f"Skewness = {skewness:.2f}")
plt.text(3, 60, f"Kurtosis = {kurt:.2f}")
plt.show()
```

```python
data = np.random.gamma(1,3,1000)

# calculate the skewness and kurtosis
skewness = skew(data)
kurt = kurtosis(data)

# plot the distribution
sns.histplot(data, kde=True)
plt.title("Distribution of Data")

# add labels to the plot
plt.text(15, 170, f"Skewness = {skewness:.2f}")
plt.text(15, 155, f"Kurtosis = {kurt:.2f}")
plt.show()
```

```python
data = np.random.gamma(2, 1, 1000) * -1

# calculate the skewness and kurtosis
skewness = skew(data)
kurt = kurtosis(data)

# plot the distribution
sns.histplot(data, kde=True)
plt.title("Distribution of Data")

# add labels to the plot
plt.text(0, 170, f"Skewness = {skewness:.2f}")
plt.text(0, 155, f"Kurtosis = {kurt:.2f}")
plt.show()
```
